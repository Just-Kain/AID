```python
import difflib
import numpy as np
def compare_files(file1_path, file2_path):
    """Сравнивает два файла и выводит различия в стиле, похожем на diff."""
    with open(file1_path, 'r', encoding='utf-8') as f1, \
         open(file2_path, 'r', encoding='utf-8') as f2:
        lines1 = f1.readlines()
        lines2 = f2.readlines()

    # Используем Differ для построчного сравнения
    differ = difflib.Differ()
    diff = list(differ.compare(lines1, lines2))

    for s in diff:
        print(s, end='')
    count = 0
    for i in diff:
        if i[0] != ' ':
            if i[0] == '?':
                count -= 1
            else:
                count += 1
    print(count)
```

```python
content = """
#include <bits/stdc++.h>

using namespace std;

int main()
{
    ios::sync_with_stdio(false);
    cin.tie(nullptr);
    cout.tie(nullptr);

    int L;
    cin >> L;
    string st;
    cin >> st;

    bitset<40> bb(st);
    unordered_map<long long, int> dp;
    unordered_set<long long> used;

    auto bfs = [&](long long v)
    {
        int res = L;
        queue<long long> q;
        q.push(v);
        used.insert(v);
        while (!q.empty())
        {
            auto from = q.front();
            q.pop();
            bitset<40> b(from);
            res = min(res, (int)b.count());
            if (res == 1) return res;
            for (int i = 1; i < L; i++)
            {
                if (b[i] && b[i - 1])
                {
                    if (i - 2 >= 0 && !b[i - 2])
                    {
                        b[i - 2] = 1;
                        b[i - 1] = 0;
                        b[i] = 0;
                        long long nxt = b.to_ullong();
                        if (!used.count(nxt))
                        {
                            q.push(nxt);
                            used.insert(nxt);
                        }
                        b[i - 2] = 0;
                        b[i - 1] = 1;
                        b[i] = 1;
                    }
                    if (i + 1 < L && !b[i + 1])
                    {
                        b[i + 1] = 1;
                        b[i] = 0;
                        b[i - 1] = 0;
                        long long nxt = b.to_ullong();
                        if (!used.count(nxt))
                        {
                            q.push(nxt);
                            used.insert(nxt);
                        }
                        b[i + 1] = 0;
                        b[i] = 1;
                        b[i - 1] = 1;
                    }
                }
            }
        }
        return res;
    };


    cout << bfs(bb.to_ullong()) << endl;

    return 0;
}

"""
```
